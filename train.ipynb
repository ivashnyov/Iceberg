# Hockey players classification traning

```python
from resizing import image_resizing
from model import create_model, create_generators, train_model, plot_history, evaluate_model, save_model
import warnings
warnings.filterwarnings('ignore')
```

Define size to resize and folder which stores folders with images of each class

```python
scale_size = (128, 128)
data_folder = 'data'
```

Resize images

```python
image_resizing(data=data_folder, size=scale_size)
```

Define input shape for our CNN and learning rate that should be less than ususal becasue of lack of date

```python
input_shape = (128, 128, 3)
learning_rate = 0.0001
```

Create model

```python
CNN = create_model(input_shape=input_shape, learning_rate=learning_rate)
```

Create generators

```python
train_data, val_data, eval_data = create_generators(data=data_folder, size=scale_size)
```

Train our model

```python
history = train_model(model=CNN, train=train_data, val=val_data, num_epochs=300)
```

```python
plot_history(history)
```

```python
evaluate_model(CNN, eval_data)
```

```python
save_model(CNN)
```
